```python
# importing necessary libraries 

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# loading the dataset

crop_data=pd.read_csv("/kaggle/input/cp-dataset/crop_production.csv")
crop_data
```

<b> Exploratory Data Analysis and Data Pre-processing </b> 

```python
crop_data.shape

#rows X columns
```

```python
# dataset columns
crop_data.columns
```

```python
# statistical inference of the dataset

crop_data.describe()
```

```python
# viewing unique crops available in the dataset

print(crop_data['Crop'].unique())
```

```python
crop_data.info()
```

```python
# Checking missing values of the dataset in each column
crop_data.isnull().sum()
```

```python
# Dropping missing values 
crop_data = crop_data.dropna()
crop_data
```

```python
#checking
crop_data.isnull().values.any()
```

```python
# Displaying State Names present in the dataset
crop_data.State_Name.unique()
```

```python
# Adding a new column Yield which indicates Production per unit Area. 
crop_data.loc[:, 'Yield'] = crop_data['Production'] / crop_data['Area']
crop_data.head(10)
```

<i> Visualizing the dataset : </i>

```python
sns.barplot(x=crop_data["State_Name"], y=crop_data["Production"])
plt.xticks(rotation = 90)
```

```python
sns.barplot(x=crop_data["State_Name"], y=crop_data["Yield"])
plt.xticks(rotation = 90)
```

```python
sns.barplot(x=crop_data["Crop_Year"], y=crop_data["Production"])
plt.xticks(rotation = 90)
```

```python
sns.barplot(x=crop_data["Crop_Year"], y=crop_data["Yield"])
plt.xticks(rotation = 90)
```

```python
sns.lineplot(x=crop_data["Season"], y=crop_data["Yield"])
plt.xticks(rotation = 90)
```

```python
sns.lineplot(x=crop_data["Season"], y=crop_data["Production"])
plt.xticks(rotation = 90)
```

```python
sns.barplot(x=crop_data["Crop"], y=crop_data["Yield"])
plt.xticks(rotation = 60)
```

Conclusions Obtained from the dataset:

1. Kerela is the largest producer of crop in India.
2. Production per unit area (Yield ) of Puducherry is maximum.
3. In the year 2011, crop yield was maximum.


Analyzing each type of Crop 

<B> 1. RICE </B> 

```python
rice_data = crop_data[crop_data["Crop"]=="Rice"]
rice_data
```

```python
rice_data.shape
```

```python
sns.barplot(x="Season",y="Yield",data = rice_data)
plt.xticks(rotation = 90)
```

```python
sns.barplot(x="State_Name",y="Yield",data=rice_data)
plt.xticks(rotation=90)
```

```python
sns.barplot(x="Crop_Year",y="Production",data=rice_data)
plt.xticks(rotation=90)
```

```python
sns.barplot(x="Crop_Year",y="Yield",data=rice_data)
plt.xticks(rotation=90)
```

```python
sns.jointplot(x="Area",y="Production",data=rice_data,kind="reg")
plt.xticks(rotation=90)
```

Observations : 

1. Rice yield is maximum in Rabi season.
2. Rice yield is maximum in Chandigarh.
3. Rice yield has been growing a little from the year 2009 to 2014.
